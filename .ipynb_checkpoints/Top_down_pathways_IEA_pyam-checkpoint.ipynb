```python
def plot_wlc(df, title=None):
    
    # Get a list of unique scenarios in the dataframe
    scenarios = df.scenario.unique()
        
    # Filter df by emission type 
    df_direct = df.filter(variable='Direct operational GHGE').data
    df_indirect = df.filter(variable='Indirect operational GHGE').data
    df_embodied = df.filter(variable='Embodied GHGE').data

    # Custom colors for each programmation
    direct_color = '#D00000'
    indirect_color = '#FFBA08'
    embodied_color = '#1C3144'

    years = df_direct['year']
    direct_values = df_direct['value']
    indirect_values = df_indirect['value']
    embodied_values = df_embodied['value']

    plt.figure(figsize=(10, 8))

    # Plot stacked bar chart
    plt.bar(years, direct_values, label='Direct operational GHGE', color=direct_color)
    plt.bar(years, indirect_values, label='Indirect operational GHGE', bottom=direct_values, color=indirect_color)
    plt.bar(years, embodied_values, label='Embodied GHGE', bottom=direct_values + indirect_values,
            color=embodied_color)

    # Matplotlib setting
    plt.legend(loc='upper right', fontsize=12)
    plt.title(title, fontsize=20, fontweight="bold")

    plt.xlabel("Year", fontsize=16, fontweight="bold")
    plt.ylabel("MtCO2eq", fontsize=16, fontweight="bold")

    plt.ylim(0, 170)

    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt

def plot_wlc(df, title=None):
    # Filter df by programmation
    df_direct = df.filter(variable='Direct operational GHGE').data
    df_indirect = df.filter(variable='Indirect operational GHGE').data
    df_embodied = df.filter(variable='Embodied GHGE').data

    # Custom colors for each programmation
    direct_color = '#D00000'
    indirect_color = '#FFBA08'
    embodied_color = '#1C3144'

    years = df_direct['year']
    direct_values = df_direct['value']
    indirect_values = df_indirect['value']
    embodied_values = df_embodied['value']

    # Create the figure and subplots
    fig, axes = plt.subplots(3, 1, figsize=(8, 15), sharex=True)

    # Plot stacked bar chart in the subplots
    axes[0].bar(years, direct_values, color=direct_color, label='Direct operational GHGE')
    axes[1].bar(years, indirect_values, color=indirect_color, label='Indirect operational GHGE')
    axes[2].bar(years, embodied_values, color=embodied_color, label='Embodied GHGE')

    # Matplotlib settings for each subplot
    for ax in axes:
        ax.legend(loc='upper right', fontsize=12)
        ax.set_ylabel("MtCO2eq", fontsize=16, fontweight="bold")
        ax.set_ylim(0, 170)

    # Common x-axis label and title
    axes[-1].set_xlabel("Year", fontsize=16, fontweight="bold")
    plt.suptitle(title, fontsize=20, fontweight="bold")

    # Get the handles and labels from one of the subplots
    handles, labels = axes[0].get_legend_handles_labels()

    # Create the legend on the right side
    fig.legend(handles, labels, loc='center right', fontsize=14, ncol=1, bbox_to_anchor=(1.4, 0.5))

    # Adjust the spacing between subplots and the position of the legend
    plt.subplots_adjust(hspace=0.4, bottom=0.2)

    plt.show()

# Call the function with your DataFrame
plot_wlc(iamdf, title='Your Title')
```

```python
import matplotlib.pyplot as plt

def plot_wlc(df, title=None):
    # Define emission types and colors
    emission_types = ['Direct operational GHGE', 'Indirect operational GHGE', 'Embodied GHGE']
    colors = ['#D00000', '#FFBA08', '#1C3144']

    # Get a list of unique scenarios in the dataframe
    scenarios = df.scenario

    for scenario in scenarios:
        plt.figure(figsize=(10, 8))
        scenario_data = df.filter(scenario=scenario)
        years = scenario_data['year']

        for i, emission_type in enumerate(emission_types):
            plt.bar(years, scenario_data.filter(variable=emission_type).data['value'], label=emission_type, color=colors[i])

        plt.title(f"{scenario} - {title}", fontsize=20, fontweight="bold")
        plt.xlabel("Year", fontsize=16, fontweight="bold")
        plt.ylabel("MtCO2eq", fontsize=16, fontweight="bold")
        plt.ylim(0, 170)
        plt.legend(loc='upper right', fontsize=12)
        plt.tight_layout()
        plt.show()

# Usage example
plot_wlc(iamdf, title="Emission Scenarios")
```
